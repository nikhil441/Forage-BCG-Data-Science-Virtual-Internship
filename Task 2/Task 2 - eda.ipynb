## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
def plot_histogram_by_churn(dataframe, variable, figsize=(20, 5), bins=100, include_churn=True):
    """
    Plot histogram for a specific variable, optionally separated by churn status.
    """
    fig = plt.figure(figsize=figsize)
    
    if include_churn:
        ax = sns.histplot(data=dataframe, x=variable, bins=bins, hue='churn')
    else:
        ax = sns.histplot(data=dataframe, x=variable, bins=bins)

# Specify the attributes for which histograms will be plotted
attributes_to_plot = ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var',
                       'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']

# Plot histograms for each specified attribute without considering churn status
for attribute in attributes_to_plot:
    plot_histogram_by_churn(dataframe=price_df, variable=attribute, include_churn=False)
```

# Hypothesis Investigation

Now, let's assess customers' responsiveness to price changes. If an increase in prices results in customer churn, we can infer that these customers are sensitive to price. Conversely, if there is a decrease or no change in price and customers still switch, attributing their behavior to price sensitivity becomes challenging.

Utilizing the consumption data for each company in the year 2015, we will introduce new features to quantify price sensitivity. These features will include averages over the entire year, the last 6 months, and the last 3 months.

```python
#Generate mean average datasets
average_yearly = price_df.groupby(['id']).mean().reset_index()
average_last_6_months = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
average_last_3_months = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()

#Display the head of the yearly mean dataset
average_yearly.head()
```

```python
#Rename columns for the yearly mean
average_yearly = average_yearly.rename(
columns={
"price_off_peak_var": "mean_year_price_off_peak_var",
"price_peak_var": "mean_year_price_peak_var",
"price_mid_peak_var": "mean_year_price_mid_peak_var",
"price_off_peak_fix": "mean_year_price_off_peak_fix",
"price_peak_fix": "mean_year_price_peak_fix",
"price_mid_peak_fix": "mean_year_price_mid_peak_fix"
}
)

average_yearly["mean_year_price_off_peak"] = average_yearly["mean_year_price_off_peak_var"] + average_yearly["mean_year_price_off_peak_fix"]
average_yearly["mean_year_price_peak"] = average_yearly["mean_year_price_peak_var"] + average_yearly["mean_year_price_peak_fix"]
average_yearly["mean_year_price_med_peak"] = average_yearly["mean_year_price_mid_peak_var"] + average_yearly["mean_year_price_mid_peak_fix"]

#Rename columns for the last 6 months mean
average_last_6_months = average_last_6_months.rename(
columns={
"price_off_peak_var": "mean_last_6_months_price_off_peak_var",
"price_peak_var": "mean_last_6_months_price_peak_var",
"price_mid_peak_var": "mean_last_6_months_price_mid_peak_var",
"price_off_peak_fix": "mean_last_6_months_price_off_peak_fix",
"price_peak_fix": "mean_last_6_months_price_peak_fix",
"price_mid_peak_fix": "mean_last_6_months_price_mid_peak_fix"
}
)

average_last_6_months["mean_last_6_months_price_off_peak"] = average_last_6_months["mean_last_6_months_price_off_peak_var"] + average_last_6_months["mean_last_6_months_price_off_peak_fix"]
average_last_6_months["mean_last_6_months_price_peak"] = average_last_6_months["mean_last_6_months_price_peak_var"] + average_last_6_months["mean_last_6_months_price_peak_fix"]
average_last_6_months["mean_last_6_months_price_med_peak"] = average_last_6_months["mean_last_6_months_price_mid_peak_var"] + average_last_6_months["mean_last_6_months_price_mid_peak_fix"]

#Rename columns for the last 3 months mean
average_last_3_months = average_last_3_months.rename(
columns={
"price_off_peak_var": "mean_last_3_months_price_off_peak_var",
"price_peak_var": "mean_last_3_months_price_peak_var",
"price_mid_peak_var": "mean_last_3_months_price_mid_peak_var",
"price_off_peak_fix": "mean_last_3_months_price_off_peak_fix",
"price_peak_fix": "mean_last_3_months_price_peak_fix",
"price_mid_peak_fix": "mean_last_3_months_price_mid_peak_fix"
}
)

average_last_3_months["mean_last_3_months_price_off_peak"] = average_last_3_months["mean_last_3_months_price_off_peak_var"] + average_last_3_months["mean_last_3_months_price_off_peak_fix"]
average_last_3_months["mean_last_3_months_price_peak"] = average_last_3_months["mean_last_3_months_price_peak_var"] + average_last_3_months["mean_last_3_months_price_peak_fix"]
average_last_3_months["mean_last_3_months_price_med_peak"] = average_last_3_months["mean_last_3_months_price_mid_peak_var"] + average_last_3_months["mean_last_3_months_price_mid_peak_fix"]

#Merge into one dataframe
price_features = pd.merge(average_yearly, average_last_6_months, on='id')
price_features = pd.merge(price_features, average_last_3_months, on='id')

#Display the head of the price features dataframe
price_features.head()
```

```python
#Merge the price features with the churn information
price_churn_data = pd.merge(price_features, client_df[['id', 'churn']], on='id')

#Display the head of the merged dataframe
price_churn_data.head()
```

```python
#Calculate correlation matrix
correlation_matrix = price_churn_data.corr()

#Plot the correlation heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(correlation_matrix, xticklabels=correlation_matrix.columns.values, yticklabels=correlation_matrix.columns.values, cmap='Greens', annot=True, annot_kws={'size': 10})

#Set axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#Display the heatmap
plt.show()
```

The preceding plot illustrates the correlation among price variables as well as their correlation with the churn variable. However, the correlation between churn and price variables appears to be minimal. This suggests that customers' churn is not significantly influenced by changes in pricing.

Next, we will combine the client data with the price churn data for subsequent modeling steps.

```python
#Merge client data without the 'churn' column with the price churn data
merged_churn_data = pd.merge(client_df.drop(columns=['churn']), price_churn_data, on='id')

#Display the head of the merged churn data
merged_churn_data.head()
```
